# Définir une chaîne de caractères

```python
a = 'Il fait beau'
print(a)
longueur = len(a)
print("La longeur de chaine est",longueur)
```

```python
# Extraire un caractère
c = a[3]
print(c)
```

```python
# Extraire une partie de la chaîne
sous = a[8:12]
print(sous)
```

```python
lesMots = a.split(' ')
len(lesMots)
```

# Caractères spéciaux

```python
# Echappement de la quote
print('Aujourd\'hui')
```

```python
# guillemets dans les quotes
print ("  'Oui, dit-il'   ")
```

```python
# Le \n permet de coder un saut de ligne
print('Nous pourrions peut-être : \n  Faire un tour au marché')
```

```python
#Le \t vous permet de coder une tabulation
print('Nous pourrions peut-être : \n \t  Faire un tour au marché')
```
