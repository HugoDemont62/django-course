```python
# On accède aux éléments d'une liste de la même manière que pour les chaines, et on crée des sous-listes grâce au slicing
print("Premier élément:",l[0])
print("Dernier élément:",l[-1])
print("A partir de l'indice 1:",l[1:])
print("Un pas de 2:",l[::2])
```
